```python
import torch
```

```python
torch.manual_seed(0)

K = 64
N = 64

shifter = (torch.arange(0, 8) * 4).cuda()
x_compressed_4bit = torch.randint(-2**31, 2**31, (K // 8, N), dtype=torch.int32, device="cuda")
# print("x_compressed_4bit", x_compressed_4bit)

block_00 = x_compressed_4bit[0,0]
values_block_00 = (block_00 >> shifter) & 0xF
print("values_block_00", values_block_00)

patch_size = (32, 16)

loaded_patch = x_compressed_4bit[:patch_size[0] // 8, :patch_size[1]]
print("loaded_patch", loaded_patch.shape, loaded_patch)

x_decompressed = (loaded_patch[:, None, :] >> shifter[None, :, None]) & 0xF

x_decompressed.reshape(patch_size)
```
